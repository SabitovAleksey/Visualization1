## Требуется написать Python скрипт, где на вход подается:

- один из типов распределения (нормальное, равномерное, биномиальное, экспоненциальное) + параметры этого распределения (у нормального матожидание и дисперсия, у равномерного отрезок и тд)
- целое число N
## Вывод программы:

- визуализация плотности (или функции вероятности) распределения
- гистограмма из выборки N элементов этого распределения
- Три числа:
    * теоретическое матожидание
    * среднее арифметическое из N элементов
    * разница между этими числамм

```python
from scipy.stats import uniform
from scipy.stats import binom
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon
```

```python
#Нормальное распредеделение
def Normal(mu, sigma, N):
    data_norm = np.random.normal(mu, sigma, N)
    count, bins, ignored = plt.hist(data_norm, 30, density=True)

    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
    plt.grid()
    plt.title("Нормальное распределение")
    plt.ylabel('Плотность')
    plt.xlabel('X')
    plt.show()

    print("Мат. ожидание:", mu)
    print("Среднее арифметическое: ", round(data_norm.mean(), 4))
    print("Разница |mu-mean|: ", round(abs(mu-data_norm.mean()),4))
    
#Равномерное распределение
def Uniform(a, b, N):
    data_U = uniform.rvs(size = N, loc = a, scale=b-a)
    grid = np.linspace(a-1, b+1, N)
    #Plotting the results
    
    plt.hist(data_U, 30, density=True)
    plt.plot(grid, uniform(loc = a, scale=b-a).pdf(grid),
         linewidth=2, color='r')
    plt.title("Равномерное распределение")
    plt.ylabel('Плотность')
    plt.xlabel('X')
    plt.grid()
    plt.show()
    
    print("Мат. ожидание:", (a+b)/2)
    print("Среднее арифметическое: ", round(data_U.mean(),4))
    print("Разница |mu-mean|: ", round(abs((a+b)/2-data_U.mean()), 4))
    
#Биномиальное распределение    
def Binomial(n, p, N):
    grid = range(0, n+1) 
    data_binom = binom.rvs(n, p, loc = 0, size = N)

    plt.plot(grid, binom.pmf(grid,n,p), '-ok',color = 'red')
    plt.hist(data_binom, density=True)
    plt.title("Биномиальное распределение")
    plt.ylabel('Плотность')
    plt.xlabel('n')
    plt.grid()
    plt.show()
    #Вычисляем мат. ожидание
    mu = n*p

    print("Мат. ожидание:", mu)
    print("Среднее арифметическое: ", round(data_binom.mean(), 4))
    print("Разница |mu-mean|: ", round(abs(mu-data_binom.mean()),4))
    
#Экспонециальное распределение    
def Exponential(lambda_exp, N):
    grid = np.linspace(0.0001, 10/lambda_exp, N )
    data_exp = expon.rvs(scale = 1/lambda_exp, size = N)

    plt.plot(grid, expon.pdf(grid, scale = 1/lambda_exp), color = 'red')
    plt.hist(data_exp, bins = 30, density = True)
    plt.title("Нормальное распределение")
    plt.ylabel('Плотность')
    plt.xlabel('X')
    plt.grid()
    plt.show()
    #Вычисляем мат. ожидание
    mu = round(1/lambda_exp,4)

    print("Мат. ожидание:", mu)
    print("Среднее арифметическое: ", round(data_exp.mean(), 4))
    print("Разница |mu-mean|: ", round(abs(mu-data_exp.mean()),4))
    
distribution = input("Normal|Uniform|Binomial|Exponential| :" )
n = 0
if distribution == 'Normal':
    mu = float(input('mu = '))
    sigma = float(input('sigma = '))
    N = int(input('N = '))
    Normal(mu, sigma, N)
    n = 1

if distribution == 'Uniform':
    a = int(input("Введите a: "))
    b = int(input("Введите b: "))
    N = int(input("Введите N: "))
    Uniform(a, b, N)
    n = 2
    
if distribution == 'Binomial':
    n = int(input("Введите n: "))
    p = float(input("Введите p: "))
    N = int(input("Введите N: "))
    Binomial(n, p, N)
    n = 3
    
if distribution == 'Exponential':
    lambda_exp = float(input("lambda = "))
    N = int(input("Введите N: "))
    Exponential(lambda_exp, N)
    n = 4
    
if n == 0:
    print("Некорректное название распределения ")
```
